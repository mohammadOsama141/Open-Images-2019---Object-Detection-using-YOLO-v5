### Further reducing the dataframe by removing unneccossery columns 

```python
columnsToKeep= ['ImageID', 'XMin', 'XMax', 'YMin', 'YMax', 'LabelName_y']


final_dfV2 = final_df[columnsToKeep]

#renaming the LabelName_y column to 'ClassName' for clarity
final_dfV2.rename(columns={'LabelName_y': 'ClassName'}, inplace=True)
final_dfV2.head(15)
```

### adding a new column of images paths to the df

```python
image_base_path = 'C:\\Users\\PC\\OneDrive\\Desktop\\cv assign3 data\\trainsubset\\'
image_extension = '.jpg'

final_dfV2['ImagePath'] = image_base_path + final_dfV2['ImageID'] + image_extension

final_dfV2.head()
```

### preparing the dataset more

```python
#Converting bounding box coordinates
final_dfV2['center_x'] = (final_dfV2['XMin'] + final_dfV2['XMax']) / 2
final_dfV2['center_y'] = (final_dfV2['YMin'] + final_dfV2['YMax']) / 2
final_dfV2['width'] = final_dfV2['XMax'] - final_dfV2['XMin']
final_dfV2['height'] = final_dfV2['YMax'] - final_dfV2['YMin']

#Drop the old bounding box columns
final_dfV2.drop(['XMin', 'XMax', 'YMin', 'YMax'], axis=1, inplace=True)
```

```python
# Create a mapping for class labels to integers
unique_labels = final_dfV2['ClassName'].unique()
label_to_int = {label: i for i, label in enumerate(unique_labels)}

# Map the class labels to integers
final_dfV2['class_int'] = final_dfV2['ClassName'].map(label_to_int)


final_dfV2.head()
```
